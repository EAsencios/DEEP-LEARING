```python
import pandas as pd
```

```python
poke_df = pd.read_csv('/content/sample_data/pokemon.csv')
poke_df
```

```python
atk_def = poke_df[['Attack', 'Defense']]
atk_def.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
res = pf.fit_transform(atk_def)
res
```

```python
pd.DataFrame(pf.powers_, columns=['Attack_degree', 'Defense_degree'])
```

```python
intr_features = pd.DataFrame(res, columns=['Attack', 'Defense', 'Attack2', 'Attack x Defense', 'Defense2'])
intr_features
```

```python
new_df = pd.DataFrame([[95, 75], [121, 120], [77, 60]], columns=['Attack', 'Defense'])
new_df
```

```python
new_res = pf.transform(new_df)
new_intr_feature = pd.DataFrame(new_res, columns=['Attack', 'Defense', 'Attack2', 'Attack x Defense', 'Defense2'])
new_intr_feature
```
